<h3>Introducción</h3>

Los métodos de K vecinos mas cercanos (kNN por su nombre en inglés, K neareast neighbours) es un caso de aprendizaje supervisado, donde se genera un modelo a partir de instancias de entrenamiento, que luego permite predecir clasificaciones nuevas observaciones. Los métodos kNN persisten todos los datos de entrenamiento (a diferencia, por ejemplo, de los árboles de decición podados). 

La idea básica para para K=1 es, dada una nueva observacion <b>y</b>, encontrar la instancia de entrenamiento que <b>xi</b> que poseea menor distancia a <b>y</b>, D(X,Y) (el vecino mas cercano). De forma similar, para K>1, se toman las distancias a los K vecinos mas cercanos:

$$ \sum_{i=1}^{K} \frac{D(x_{i}, Y)} {n}  $$

De esta manera, al observar una nueva instancia es posible predecir la clase a la que pertenece utilizando los vecinos mas cercanos como se muestra en la siguiente figura:

<img src="files/img/kNN.png" />

Veremos en este ejercicio el efecto sobre el modelo constrido al tomar diferentes tamaños de vecindad (tomar distintos valores de K). Al igual que en el ejercicio anterior, nos centraremos en el efecto de sobreajuste a los datos de entrenamiento.
También estudiaremos diferentes funciones para calcular la distancia.




<h3>Datos utilizados para los experimentos</h3>
En los experimentos de las secciones siguientes se utilizara el conjunto de datos de <a href="http://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes"> Pima Indians Diabetes </a> del repositorio de <a href="https://archive.ics.uci.edu/ml/index.html">UCI</a>.

Este conjunto de datos cuenta con 768 observaciones, donde cada una cuenta con los siguientes 8 atributos:
<pre>
   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)
</pre>

La clase de cada observación es binaria: tiene diabetes (1) o no (0). Se cuenta con 500 observaciones de pacientes sin diabetes (clase 0) y 268 de pacientes con diabetes (clase 1).

<h3>Set de datos, instancias de test y entrenamiento</h3>
El primer paso es cargar el set de datos (el archivo CSV de diabetes de la UCI). Luego, se utiliza numpy para separar los atributos (8 atributos en este caso) de las clases.

```python
import urllib
import numpy as np
from sklearn import cross_validation
from sklearn.neighbors import KNeighborsClassifier

# URL for the Pima Indians Diabetes dataset (UCI Machine Learning Repository)
raw_data = open("data/pima-indians-diabetes.data.csv", "r") #Downloaded from http://goo.gl/j0Rvxq

# load the CSV file as a numpy matrix
dataset = np.loadtxt(raw_data, delimiter=",")

# Separate the data from the target attributes
X = dataset[:,0:8] 	# primeras 8
y = dataset[:,8] 	# objetivo

print "Atributos:", X.shape
print "Clases:", y.shape
```

A continuación se utiliza el módulo 'cross validation' para separar en dos conjuntos de 'entrenamiento' y 'prueba' en una proporción 80-20 respectivamente. Se utilizaran entonces 614 instacias para entrenar el modelo y 154 instancias para poner a prueba su correctitud.

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2, random_state=0)
print "Instancias de Entrenamiento:", y_train.shape
print "Instancias de Prueba:", y_test.shape
```

<h3> IBK segun el tamaño de la vecindad</h3>
A continuacion utilizaremos clasificadores de vecindad mas cercana, variando el tamaño considerado para la vecidad. Para esto, al crear el clasificador utilizamos el parametro n_neighbors. En este caso se utiliza por defecto pesos uniformes, por lo que todos los puntos en cada vecidad sos pesados igualmente.

```python
xaxis = []
yserie_train = []
yserie_test = []
testRange = range(1,200, 2)
for i in testRange:
	# Create the model and train it
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, y_train)
	
    # Test using the train and test data
    xaxis.append(i)
    yserie_train.append(neigh.score(X_train, y_train))
    yserie_test.append(neigh.score(X_test, y_test))
```

El siguiente gráfico muestra el score para los datos de entrenamiento (en verde) y para los datos de prueba (en azul) al variar la cantidad de vecinos considerada.

<b>NOTA:</b> Si se desea ver con mayor detalle una determina area, es posible cambiar parametro sizeRange y ejecutar la celda del notebook para regenerar el gráfico

```python
%pylab inline 
plot(xaxis, yserie_train, color='g', label="Datos de Entrenamiento")
plot(xaxis, yserie_test, color='b', label="Datos de Prueba")
grid('on')
xlabel('Tamano de la vecidad')
ylabel('Score')
title('Score para vecinos mas cercanos')
legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
#
yserie_distance = []
for i in testRange:
	neigh2 = KNeighborsClassifier(n_neighbors=i, weights='distance')
    neigh2.fit(X_train, y_train)
    
    yserie_distance.append(neigh2.score(X_test, y_test) )
#
```
